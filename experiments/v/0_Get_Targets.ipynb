```python
import pandas as pd
from pathlib import Path
import zipfile

# Extract all zip files in 'data/btc' folder
data_path = Path("../../data/btc")
i = 0
for file in data_path.glob("*.zip"):
    with zipfile.ZipFile(file, "r") as zip_ref:
        zip_ref.extractall(data_path)
    i += 1

# Read all CSV files and concatenate into a single DataFrame
csv_files = list(data_path.glob("bitstamp_trx*.csv"))
df = pd.concat((pd.read_csv(csv_file) for csv_file in csv_files), ignore_index=True)
df = df.rename(columns={c: c.split("'")[1] for c in df.columns})

# Display the head of the resulting DataFrame
df.head()
```

```python
df["datetime"] = pd.to_datetime(df["datetime"])
df = df.set_index("datetime")
df.sort_index()
```

```python
returns = df["price"].sort_index().pct_change()
returns = returns.dropna()
returns
```

```python
vol = (returns**2).resample("h").sum().rename("vol")
ret = returns.add(1).resample("h").prod().add(-1).rename("ret")
hourly = pd.merge(vol, ret, left_index=True, right_index=True, how="outer")
hourly
```

```python
merged_df = hourly.copy()
merged_features = []

for target_index in hourly.index:
    segment = returns.loc[target_index - pd.Timedelta("60m") : target_index]
    print(segment.index.min(), segment.index.max(), target_index)
    print(segment.shape, segment.to_numpy().flatten().shape)
    segment = segment.add(1).resample("5min").prod().add(-1).rename("ret")
    # if segment.shape[0] != 120:
    #     continue
    merged_features.append([target_index, *segment.to_numpy().flatten().tolist()])
```

```python
hourly.index.min()
```

```python
merged_targets_df = pd.DataFrame(merged_features)
merged_targets_df = merged_targets_df.rename(columns={0: "datetime"})
merged_targets_df = merged_targets_df.set_index("datetime")
merged_targets_df = merged_targets_df.iloc[:, :-1]
merged_targets_df.columns = [f"ret_t-{i}" for i in range(1, 13)]
merged_targets_df = merged_targets_df.iloc[1:]
merged_targets_df
```

```python
final_df = pd.merge(
    hourly, merged_targets_df, left_index=True, right_index=True, how="outer"
)
final_df = final_df.iloc[1:]
final_df
```
