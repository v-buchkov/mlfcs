```python
import pandas as pd
from pathlib import Path

data_path = Path("../../data/btc")

features = pd.read_csv(data_path / "features.csv")
targets = pd.read_csv(data_path / "targets.csv")

features.shape, targets.shape
```

```python
data_df = pd.read_csv(data_path / "data_df.csv")
data_df["datetime"] = pd.to_datetime(data_df["datetime"])
data_df = data_df.set_index("datetime")
data_df.head()
```

```python
data_df = data_df.dropna(axis=0)
data_df.shape
```

```python
data_df.isna().any().any()
```

```python
data_df["datetime"] = pd.to_datetime(data_df["datetime"])
data_df["datetime"].min(), data_df["datetime"].max()
```

```python
features["time"] = pd.to_datetime(features["time"])
features = features.set_index("time")
```

```python
targets["datetime"] = pd.to_datetime(targets["datetime"])
targets = targets.set_index("datetime")
```

```python
features.sort_index(ascending=True)
```

```python
targets.index = targets.index.tz_localize(None)
targets = targets.loc[features.index.min() :]
targets
```

```python
features.index.max(), targets.index.max()
```

```python
features = features.loc[: targets.index.max() - pd.Timedelta("1s")]
```

```python
merged_df = targets.copy()
merged_features = []

for target_index in targets.index:
    segment = features.loc[target_index - pd.Timedelta("60m") : target_index]
    print(segment.index.min(), segment.index.max(), target_index)
    print(segment.shape, segment.to_numpy().flatten().shape)
    if segment.shape[0] != 120:
        continue
    merged_features.append([target_index, *segment.to_numpy().flatten().tolist()])
```
