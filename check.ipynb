```python
import pandas as pd

w = pd.read_csv("weibull_online_monthly.csv")
w["date"] = pd.to_datetime(w["date"])
w = w.set_index("date")
w.head()
```

```python
true = pd.read_csv("data/btc/data_df.csv")
true["datetime"] = pd.to_datetime(true["datetime"])
true = true.set_index("datetime")
true = true["vol"]
true.head()
```

```python
m = pd.read_csv("output_retrain.csv")
m["datetime"] = pd.to_datetime(m["datetime"])
m = m.set_index("datetime")
m
```

```python
d = m.merge(w, left_index=True, right_index=True, how="left")
d = d.merge(true, left_index=True, right_index=True, how="left")
d.shape
```

```python
d = d.dropna(subset=["vol", "SigmaLSTMPredictor"])
d
```

```python
from sklearn.metrics import root_mean_squared_error

root_mean_squared_error(d["vol"], d["LSTMSoftplusPredictor"])
```
